```python
df= pd.read_csv("mushrooms.csv")
df
```

```python
df.info()
```

```python
df.describe().T
```

```python
y = df.classes
y
```

```python
X = pd.get_dummies(df.drop('classes', axis=1))
y = df['classes'].apply(lambda x: 1 if x == 'p' else 0)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
print(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn5 = KNeighborsClassifier(n_neighbors = 5)
knn1 = KNeighborsClassifier(n_neighbors = 1)
knn5.fit(X_train, y_train)
knn1.fit(X_train, y_train)

y_pred_5 = knn5.predict(X_test)
y_pred_1 = knn1.predict(X_test)
print(f"Accuracy with k=5: {accuracy_score(y_test, y_pred_5) * 100:.2f}%")
print(f"Accuracy with k=1: {accuracy_score(y_test, y_pred_1) * 100:.2f}%")
```

```python
from sklearn.metrics import precision_score, recall_score,f1_score

precision5 = precision_score(y_test,y_pred_5)
precision1 = precision_score(y_test,y_pred_1)

recall5 = recall_score(y_test,y_pred_5)
recall1 = recall_score(y_test,y_pred_1)

f1_5 = f1_score(y_test, y_pred_5)
f1_1 = f1_score(y_test, y_pred_1)

print(f'Precision5: {precision5 * 100:.2f}%')
print(f'Recall5: {recall5 * 100:.2f}%')
print(f'F1 Score: {f1_5 * 100:.2f}%')

print(f'Precision1: {precision1 * 100:.2f}%')
print(f'Recall1: {recall1 * 100:.2f}%')
print(f'F1 Score: {f1_1 * 100:.2f}%')
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm1 = confusion_matrix(y_test, y_pred_5)
print("Confusion Matrix:\n", cm1)

# Plot the confusion matrix
plt.figure(figsize=(10,7))
sns.heatmap(cm1, annot=True, fmt='d', cmap='Blues', xticklabels=['Poisonous', 'Edible'], yticklabels=['Poisonous', 'Edible'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for KNN(k=5) Classifier')
plt.show()
```

```python
cm2 = confusion_matrix(y_test, y_pred_1)
print("Confusion Matrix:\n", cm2)

# Plot the confusion matrix
plt.figure(figsize=(10,7))
sns.heatmap(cm2, annot=True, fmt='d', cmap='Blues', xticklabels=['Poisonous', 'Edible'], yticklabels=['Poisonous', 'Edible'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for KNN(k=1) Classifier')
plt.show()
```

```python
import matplotlib.pyplot as plt
edible_count = df['classes'].value_counts()['e']
poisonous_count = df['classes'].value_counts()['p']
total_count = len(df)

percentage_edible = (edible_count / total_count) * 100
percentage_poisonous = (poisonous_count / total_count) * 100

print("Percentage of edible mushrooms: {:.2f}%".format(percentage_edible))
print("Percentage of poisonous mushrooms: {:.2f}%".format(percentage_poisonous))


plt.figure(figsize=(8, 6))
plt.bar(['Edible', 'Poisonous'], [percentage_edible, percentage_poisonous], color=['green', 'red'])
plt.xlabel('Mushroom Type')
plt.ylabel('Percentage')
plt.title('Percentage of Edible and Poisonous Mushrooms')
plt.show()

```
